This is just an example of what the data looks like. <br>

# Splitting

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.12, random_state=42)
for train_index, test_index in split.split(df, df['target']):
    strat_train = df.loc[train_index]
    strat_test = df.loc[test_index]
strat_train['target'].value_counts()
```

```python
strat_test['target'].value_counts()
```

```python
#Scaling and imputing pipeline
data_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy = 'median')),
    ('std_scaler', StandardScaler()),
])

#Numpy array output
data_num = data_pipe.fit_transform(data)
data_num.shape
```
